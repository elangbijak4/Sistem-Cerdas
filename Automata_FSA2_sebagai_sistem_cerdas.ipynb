```python
import yfinance as yf
import time

# FSA States
class FSA:
    def __init__(self):
        self.state = 'INIT'

    def transition(self, signal):
        if self.state == 'INIT':
            if signal == 'BUY':
                self.state = 'BUYING'
            elif signal == 'HOLD':
                self.state = 'HOLDING'
            elif signal == 'SELL':
                self.state = 'SELLING'

        elif self.state == 'BUYING':
            if signal == 'SELL':
                self.state = 'SELLING'
            elif signal == 'HOLD':
                self.state = 'HOLDING'

        elif self.state == 'SELLING':
            if signal == 'BUY':
                self.state = 'BUYING'
            elif signal == 'HOLD':
                self.state = 'HOLDING'

        elif self.state == 'HOLDING':
            if signal == 'SELL':
                self.state = 'SELLING'
            elif signal == 'BUY':
                self.state = 'BUYING'

        return self.state

# Decision making based on stock price movement
def make_decision(previous_price, current_price):
    if current_price > previous_price:
        return 'BUY'
    elif current_price < previous_price:
        return 'SELL'
    else:
        return 'HOLD'

# Main loop for real-time stock price monitoring
def stock_trading_bot(stock_symbol, interval=5):
    fsa = FSA()
    previous_price = None

    while True:
        # Get real-time stock data
        stock = yf.Ticker(stock_symbol)
        current_price = stock.history(period='1d')['Close'][-1]

        if previous_price is not None:
            decision = make_decision(previous_price, current_price)
            current_state = fsa.transition(decision)

            print(f"Current Price: {current_price}, Decision: {decision}, State: {current_state}")

        previous_price = current_price
        time.sleep(interval)

# Start trading bot for stock symbol
stock_trading_bot('AAPL', interval=10)
```
